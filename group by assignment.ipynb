# `Overall revenue insights`


```python
product_revenue = df.groupby('Product')['Revenue'].sum()

print("Total Revenue for Each Product:")
pd.DataFrame(product_revenue)
```

 # `Average revenue patterns`

```python
average_revenue_per_product = df.groupby('Product')['Revenue'].mean()

print("Average Revenue per Product:")
pd.DataFrame(average_revenue_per_product)
```

```python
df['Date'] = pd.to_datetime(df['Date'])

average_revenue_by_date = df.groupby('Date')['Revenue'].mean()

print("Average Revenue Patterns Over Time:")
pd.DataFrame(average_revenue_by_date)
```

# `Peak quantity sold observations`



```python
peak_quantity_observation=df.groupby("Product")["Quantity"].max()

print("Peak Quantity Sold Observations:")
pd.DataFrame(peak_quantity_observation)
```

# `Minimum quantity sold occurrences`


```python
min_quantity_occurrences =df.groupby("Product")["Quantity"].min()
pd.DataFrame(min_quantity_occurrences)
```

## Aggregate DataFrame df by extracting the month and year from the 'Date' column. Calculate the collective revenue for each unique combination of month and year.

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year
```

```python
df[['Date','Month','Year']]
```

```python
date_month=df.groupby(["Month","Year"])["Revenue"].sum()
pd.DataFrame(date_month)
```

## Group the DataFrame df by the 'Date' column and find the date with the highest total revenue and the date with the highest total quantity sold.

```python
 df['Date'] = pd.to_datetime(df['Date'])

daily_totals = df.groupby('Date')[["Revenue","Quantity"]].agg((np.sum))

max_revenue_date = daily_totals.loc[daily_totals['Revenue'].idxmax()]        

max_quantity_date = daily_totals.loc[daily_totals['Quantity'].idxmax()]          

```

```python
df['Date'] = pd.to_datetime(df['Date'])

daily_totals = df.groupby('Date')[["Revenue","Quantity"]].agg((np.sum))

print("Date with the Highest Total Revenue:")
pd.DataFrame(max_revenue_date).T

```

```python
df['Date'] = pd.to_datetime(df['Date'])

daily_totals = df.groupby(['Date'])[["Revenue","Quantity"]].agg((np.sum))


print("Date with the Highest Total Quantity Sold:")
pd.DataFrame(max_quantity_date).T
```

## Group the DataFrame df by the 'Product' column andadd a new column 'Revenue Percentage' that shows the percentage of each product's revenue relative to the total revenue for that product.

```python
df['Revenue_total'] = df.groupby('Product')['Revenue'].transform('sum')

df['Revenue Percentage'] = (df['Revenue'] / df['Revenue_total']) * 100

df = df.drop('Revenue_total', axis=1)

pd.DataFrame(df)
```
