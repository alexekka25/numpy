## Using .loc method, find the number of units of Smart Tv sold in Central Channai in year 2020

## Give a detailed data about the Tamilnadu, and then tell me how many items were sold in total from this warehouse over the 10 year period?

```python
tamilnadu_data = df.xs(key='Tamilnadu', axis=0, level='State')

print("Detailed Data for Tamilnadu:")
print(tamilnadu_data)

total_items_sold = tamilnadu_data.sum().sum()

print(f"\nTotal items sold from Tamilnadu over the 10-year period: {total_items_sold}")
total_items_sold
```

## How many Smart TV were sold in Telangana & Karnataka during the year 2017?

```python
smart_tv_sales_2017 = df.xs(key='Smart TV', level='Product').loc[:, '2017']

smart_tv_sales_telangana = smart_tv_sales_2017.xs(key='Telangana', level='State').sum()
smart_tv_sales_karnataka = smart_tv_sales_2017.xs(key='Karnataka', level='State').sum()

x=smart_tv_sales_telangana.sum()
x1=smart_tv_sales_karnataka.sum()

print(f"Number of Smart TVs sold in Telangana during 2017: {smart_tv_sales_telangana}")
print(f"Number of Smart TVs sold in Karnataka during 2017: {smart_tv_sales_karnataka}")

print(f"total Smart TVs sold in Telangana during 2017: {x}")
print(f" total Smart TVs sold in Karnataka during 2017: {x1}")
```

##  What method is used to know if the MultiIndex index is sorted or not

## 1st method

```python
is_sorted = df.index._is_lexsorted()

print(f" the MultiIndex lexsorted: {is_sorted}")
```

## 2nd method

```python
is_sorted = all(df.index[i] <= df.index[i + 1] for i in range(len(df.index) - 1))
print("Is the MultiIndex sorted?", is_sorted)
```
